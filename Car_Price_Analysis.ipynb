```python
# ðŸš— Car Price Analysis â€” IBM Data Analysis Lab
# Author: Denis Naumov
# GitHub: https://github.com/DenisNaumov7777
# Description:
# Exploratory Data Analysis (EDA) of automobile dataset to identify 
# variables that significantly affect car prices.

# -------------------------------------------------------------
# 1ï¸âƒ£ Import Required Libraries
# -------------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configure display
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
sns.set(style="whitegrid")
%matplotlib inline

# -------------------------------------------------------------
# 2ï¸âƒ£ Load Dataset
# -------------------------------------------------------------
# Dataset link provided by IBM Developer Skills Network
file_path = "data/automobileEDA.csv"
df = pd.read_csv(file_path)
print("âœ… Local dataset loaded successfully.")
df.head()

# -------------------------------------------------------------
# 3ï¸âƒ£ Data Types Overview
# -------------------------------------------------------------
print("Data types of each column:")
print(df.dtypes)

# Example: Check a specific column
df['peak-rpm'].dtypes

# -------------------------------------------------------------
# 4ï¸âƒ£ Correlation Between Numerical Variables
# -------------------------------------------------------------
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])
numeric_df.corr()

# Focus on specific numeric subset
df[['bore', 'stroke', 'compression-ratio', 'horsepower']].corr()

# -------------------------------------------------------------
# 5ï¸âƒ£ Visualize Relationships Using Regression Plots
# -------------------------------------------------------------
# Engine size vs Price
sns.regplot(x="engine-size", y="price", data=df)
plt.title("Engine Size vs Price")
plt.ylim(0,)
plt.show()

# Highway-mpg vs Price
sns.regplot(x="highway-mpg", y="price", data=df)
plt.title("Highway MPG vs Price")
plt.ylim(0,)
plt.show()

# Peak RPM vs Price
sns.regplot(x="peak-rpm", y="price", data=df)
plt.title("Peak RPM vs Price")
plt.show()

# Stroke vs Price
sns.regplot(x="stroke", y="price", data=df)
plt.title("Stroke vs Price")
plt.show()

# -------------------------------------------------------------
# 6ï¸âƒ£ Boxplots for Categorical Variables
# -------------------------------------------------------------
sns.boxplot(x="body-style", y="price", data=df)
plt.title("Body Style vs Price")
plt.show()

sns.boxplot(x="engine-location", y="price", data=df)
plt.title("Engine Location vs Price")
plt.show()

sns.boxplot(x="drive-wheels", y="price", data=df)
plt.title("Drive Wheels vs Price")
plt.show()

# -------------------------------------------------------------
# 7ï¸âƒ£ Descriptive Statistics
# -------------------------------------------------------------
print("Descriptive statistics for numeric variables:")
display(df.describe())

print("Descriptive statistics for categorical variables:")
display(df.describe(include=['object']))

# -------------------------------------------------------------
# 8ï¸âƒ£ Value Counts for Key Categorical Features
# -------------------------------------------------------------
drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.reset_index(inplace=True)
drive_wheels_counts.rename(columns={'index': 'drive-wheels', 'drive-wheels': 'value_counts'}, inplace=True)
drive_wheels_counts

engine_loc_counts = df['engine-location'].value_counts().to_frame()
engine_loc_counts.rename(columns={'engine-location': 'value_counts'}, inplace=True)
engine_loc_counts.index.name = 'engine-location'
engine_loc_counts.head(10)

# -------------------------------------------------------------
# 9ï¸âƒ£ Grouping and Pivot Table Analysis
# -------------------------------------------------------------
# Group by drive-wheels
df_group_one = df[['drive-wheels', 'body-style', 'price']]
df_grouped = df_group_one.groupby(['drive-wheels'], as_index=False).agg({'price': 'mean'})
print("Average price by drive-wheel type:")
display(df_grouped)

# Group by drive-wheels and body-style
df_gptest = df[['drive-wheels', 'body-style', 'price']]
grouped_test1 = df_gptest.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
grouped_pivot = grouped_test1.pivot(index='drive-wheels', columns='body-style')
grouped_pivot = grouped_pivot.fillna(0)
display(grouped_pivot)

# -------------------------------------------------------------
# ðŸ”Ÿ Heatmap Visualization
# -------------------------------------------------------------
plt.figure(figsize=(8,6))
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.title("Average Car Price by Drive-wheels and Body-style")
plt.show()

# Alternative annotated heatmap
fig, ax = plt.subplots(figsize=(8,6))
im = ax.pcolor(grouped_pivot, cmap='RdBu')

# Label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)
plt.xticks(rotation=90)
fig.colorbar(im)
plt.title("Pivot Heatmap â€” Drive-wheels vs Body-style")
plt.show()

# -------------------------------------------------------------
# 1ï¸âƒ£1ï¸âƒ£ Pearson Correlation and Statistical Significance
# -------------------------------------------------------------
def corr_test(var):
    pearson_coef, p_value = stats.pearsonr(df[var], df['price'])
    print(f"{var:15s} â†’ Pearson: {pearson_coef:.3f},  P-value: {p_value:.4f}")

vars_to_test = [
    'wheel-base', 'horsepower', 'length', 'width',
    'curb-weight', 'engine-size', 'bore',
    'city-mpg', 'highway-mpg'
]

print("Pearson correlation coefficients with price:\n")
for v in vars_to_test:
    corr_test(v)

# -------------------------------------------------------------
# âœ… Conclusion
# -------------------------------------------------------------
# Based on the correlation analysis and visualizations:
# - Strong positive correlation: engine-size, curb-weight, horsepower, width
# - Strong negative correlation: city-mpg, highway-mpg
# - Key categorical feature: drive-wheels
# These variables are likely the best predictors of car price for model training.
```
